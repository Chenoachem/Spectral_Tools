```python
Absorption_0 = fits.open('Beam01A_Test0_Absorption_cube.fits')
Absorption0_data = Absorption_0[0].data
Absorption0_header = Absorption_0[0].header

rp = Absorption_0[0].header['CRPIX3']
rf = Absorption_0[0].header['CRVAL3']
df = Absorption_0[0].header['CDELT3']
nf = Absorption_0[0].header['NAXIS3']
Absorption0_xvals = rf + df*(np.arange(nf)-rp)
Absorption0_xvals=Absorption0_xvals*10**-6
print(Absorption0_xvals[2000])
```

```python
ra="16:36:43.1" #16:36:42.6, -47:31:36
dec="-47:31:38"
c = SkyCoord(ra, dec, unit=(u.hourangle, u.deg))
```

```python
w = wcs.WCS(Absorption0_header, naxis=2)
xpix,ypix=c.to_pixel(w,origin=0,mode='wcs')    
xpix_0=int(xpix)
ypix_0=int(ypix)
print(xpix_0,ypix_0)
```

```python
signal_0=[]
for x in range(0, 3842):
    value = np.mean(Absorption0_data[x,ypix_0-3:ypix_0+3,xpix_0-3:xpix_0+3])
    signal_0.append(value)
print(signal_0[1368])
```

```python
Tc=np.median(signal_5[500:1000])
print(Tc)
```

```python
bigfig=plt.figure(figsize=(20,12))
ax1=bigfig.add_subplot(111)
ax1.step(Absorption0_xvals,signal_0,color='blue', label='Test0')
ax1.step(Absorption1_xvals,signal_1,color='red', label='Test1')
ax1.step(Absorption2_xvals,signal_2,color='green', label='Test2')
ax1.step(Absorption3_xvals,signal_3,color='plum', label='Test3')
ax1.step(Absorption4_xvals,signal_4,color='cyan', label='Test4')
ax1.set_title("Absorption Source",fontsize=18)
ax1.set_xlabel("Frequency (MHz)",fontsize=18)
ax1.set_ylabel("Intensity (Jy)",fontsize=18)
ax1.set_xlim(1665.7,1666.0)
ax1.legend(prop={'size': 18})
ax1.tick_params(labelsize=16)
bigfig.savefig("Absorption_Source_Compare_Spectra.png")
```
